```python
import pandas as pd
```

```python
df=pd.read_csv("cleaned_data.csv")
```

```python
df.head(10)
```

now creating lag prices for future prediction


```python
df["lag_1"]=df["Modal Price (Rs./Quintal)"].shift(1)
```

```python
df
```

```python
df["lag_2"]=df["Modal Price (Rs./Quintal)"].shift(2)
df["lag_3"]=df["Modal Price (Rs./Quintal)"].shift(3)
```

```python
df
```

```python
# adding moving averages
df["ma_3"]=df["Modal Price (Rs./Quintal)"].rolling(window=3).mean() ## mean of 3 date model price consicutively
```

```python
df
```

```python
df["ma_7"]=df["Modal Price (Rs./Quintal)"].rolling(window=7).mean()
```

```python
df
```

```python
df.fillna({'lag_1': df["Modal Price (Rs./Quintal)"],'lag_2':df["Modal Price (Rs./Quintal)"],'lag_3':df["Modal Price (Rs./Quintal)"],'ma_3':df["Modal Price (Rs./Quintal)"],'ma_7':df["Modal Price (Rs./Quintal)"]},inplace=True)
```

```python
df
```

Without lag/rolling features, the model only sees "today’s price" and maybe the date. It can’t learn how the market behaves over time.
Lagging = short-term memory
Rolling = trend + stability measure



so now instand of date we can relay on new feature with smooth shifting of data

```python
df.columns
```

```python
X = df.drop(columns=["Modal Price (Rs./Quintal)", "Min Price (Rs./Quintal)", "Max Price (Rs./Quintal)", "Price Date"])
y = df["Modal Price (Rs./Quintal)"]
```

```python
X=X.drop("Grade",axis=1)
```

```python
y
```

```python
#features=X.columns
```

```python
#X=pd.get_dummies(X[features])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=False)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_lr = lr.predict(x_test)

print("Linear Regression Performance:")
print(f"MAE: {mean_absolute_error(y_test, y_pred_lr):.2f}")
print(f"RMSE: {mean_squared_error(y_test, y_pred_lr):.2f}")
print(f"R²: {r2_score(y_test, y_pred_lr):.2f}")
```

```python
rf=RandomForestRegressor(n_estimators=200,random_state=123)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred_rf=rf.predict(x_test)
print("Random Forest Performance:")
print(f"MAE: {mean_absolute_error(y_test, y_pred_rf):.2f}")
print(f"RMSE: {mean_squared_error(y_test, y_pred_rf):.2f}")
print(f"R²: {r2_score(y_test, y_pred_rf):.2f}")
```

```python
import joblib
joblib.dump(rf, "best_price_model.pkl")
```
